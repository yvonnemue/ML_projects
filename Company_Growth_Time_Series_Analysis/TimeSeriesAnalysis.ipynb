# Time Series Analysis

## Summary

### Problem: 

We have fake purchases data from a possible client. 

We have 2 years and 4 months (850 days) worth of revenue data. 

Our goal is to decide whether to invest or not in the company.

### Looking for patterns:  
How is the company evolving? Is the revenue growing?

- Trend is positive.
- Revenue is growing. But the growth is decreasing, it was huge during the first two years and seems to be stabilizing in 2021. 

Would you invest money on them? 

- Compared to 2019, in 2020 the company's revenue grew a 376%. And during the 2021 Q1 the growth was 169% .

- If we were to invest money in the company and expect the company to pay back the 6% of their revenues each month, after doing an estimation of the company's revenues for the next year, in 12 months the company would be able to pay back around €45,000 (growth of 89%). 

Is there a temporal pattern in the revenue? Why does it happen?

- There is weekly seasonality, with peaks over the weekends, especially on Saturdays. Especially in the last months, most of the revenue takes place on Satudays and Sundays.

### Insights 

+ Although the company is growing and its revenue follows a positive trend, their predicted growth is only going to be 89% in the next year. 

+ If we were to invest €100K in the company, in order to have the money returned within 1 year, the company should have a growth of 334%.

+ If, instead of €100K, we invest €50K, the growth should only be around 117%. This option is more viable. There is still a risk, because the predicted annual growth of the company for the next 12 months is only 89%. Nevertheless, if the company increases its growth above our predictions thanks to the marketing campaigns, the goal is more achievable.

### Stationarity and growth

As wee saw in the previous plots, our data seems to be non-stationary, e.g. it depends on time. Since we ultimately want to do some forecasting, we need to make sure our data is stationary with the Augmented Dickey-Fuller test.

Null hypothesis of the test is that the data is non-stationary.
If the null is rejected, then the data is stationary. (Rule of thumb for rejecting the null, p-value < 0.05)

```python
from statsmodels.tsa.stattools import adfuller
result = adfuller(data.Revenue)

print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

Since p value is way higher than 0.05 and ADF statistic is higher than all critical values, **null hypothesis is NOT rejected**, therefore our data is **non-stationary** and we should make amends.

#### Making it stationary
By taking the first differences of the data we can make it stationary.

```python
result = adfuller(data['Revenue'].diff()[1:])
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

Our data is now stationary, with p value lower than 0.05 and ADF statistic also lower than all critical values.

```python
stat_data = data['Revenue'].diff()
```

### Discovering seasonality in the data
In order to discover if there is seasonality, we can run autocorrelations.
Lighter blue shadowing shows the confidence interval.
The autocorrelation with lag zero always equals 1, because this represents the autocorrelation between each day and itself. 

```python
from statsmodels.graphics.tsaplots import plot_acf
plot_acf(data.Revenue)
plt.xlabel('Lags')
plt.ylabel('Correlation')
plt.show()
```

```python
# Relationship/autocorrelation after removing the effects of lags
from statsmodels.graphics.tsaplots import plot_pacf
plot_pacf(data.Revenue)
plt.xlabel('Lags')
plt.ylabel('Correlation')
plt.show()
```

There **is seasonality** in our data, as show by the previous plots.

## Predictions with the ARIMA model

### Let's use auto arima (Automated Hyndman & Khandakar Algorithm) to choose the best p, d and q values for the ARIMA model


p = periods to lag for eg: (if p = 5 then we will use the five previous periods of our time series in the autoregressive portion of the calculation) 

d = number of differencing transformations required by the time series to get stationary

q = lag of the error component (part of the time series not explained by trend or seasonality). The size of the moving average window.

But first let's split the data into train and test. We keep from 0 to 750 as train and from 751 to 850 as test.

Best model: ARIMA(5,1,1) = combining a 5th order Auto-Regressive model, with a differencing of level 1, and a 1st order Moving Average model.

```python
test_size = 100

df_train = data[:-test_size]
df_test = data[-test_size:]
```

```python
from pmdarima.arima import auto_arima
model = auto_arima(df_train.Revenue, trace=True, error_action='ignore', suppress_warnings=True)
print(model.summary())
```

```python
from statsmodels.tsa.arima_model import ARIMA
model = ARIMA(df_train.Revenue, order=(5,1,1))
result = model.fit()

y_arima = result.predict(start=700,end=849)
ax = result.plot_predict(start=700,end=849)
plt.show()
```

#### Comparing the prediction with the actual values

```python
# Create df with the predictions to then merge it with the actual data
y_arima = pd.DataFrame(y_arima)
```

```python
y_arima.head()
```

```python
y_real = data[700:]['Revenue']
table_arima = y_arima.merge(y_real, right_index=True, left_index=True)
table_arima.columns = ['Predicted', 'Real']
```

## Predictions with Facebook Prophet

### Preparing the df for the Prophet

```python
df_train_proph = df_train[['Date', 'Revenue']][500:700]
df_train_proph = df_train_proph.rename(columns={'Date':'ds', 'Revenue':'y'})

df_test_proph = df_test[['Date', 'Revenue']]
df_test_proph = df_test_proph.rename(columns={'Date':'ds', 'Revenue':'y'})
```

### Training the model

Regarding weekly seasonality plot with Prophet: y-axis in the weekly plot represents how much the day of week affect the prediction. Also, weekly components sum to zero, which explains the negative and positive values.

```python
from prophet import Prophet
```

```python
# Initialize and train a model
m = Prophet()

forecast = m.fit(df_train_proph)

# Make predictions
future = m.make_future_dataframe(periods=test_size, freq='d',include_history=True)
forecast = m.predict(future)
forecast.head()
m.plot(forecast);
m.plot_components(forecast);
```

```python
# Create a table with the real and predicted values
y_hat_proph = pd.DataFrame(forecast.yhat)
y_hat_proph.index = range(550,850)
table_proph = pd.DataFrame(df_test['Revenue']).merge(y_hat_proph, how='left', left_index=True, right_index=True)
table_proph = table_proph.rename(columns={'Revenue':'Real', 'yhat':'Pred'})
table_proph.head()
```

```python
# Visualization of predicted vs real data. Blue is real and orange is prediction
table_proph.plot()
```

```python
# Compute the MSE for the model and compare it with the arima
mse['mse_prophet'] = mean_squared_error(table_proph.Real, table_proph.Pred)
mse
```

Since the MSE of the Prophet model is lower, it means that it performs best. Therefore we can do our final predictions with the Prophet.

### Final predictions

In general, predictions are best when they are close in time to our values. Therefore, in order to make realiable forecasts we will train the model again with the most recent data. 
Due to the nature fo time series, we can not test this final model with unseen data. What we can do is to regularly check if there are data drifts and model drifts in the predictions.

```python
# Creating the df
data_for_pred = data[['Date', 'Revenue']][-365:]
data_for_pred = data_for_pred.rename(columns={'Date':'ds', 'Revenue':'y'})
```

### Desirable growth

#### If we lend €100K

If we lend €100K and we want to have it back before 1 year, the company should have an annual growth of 334%.

To do so we need to multiply each value per the desirable growth and the interest rate. And divide the money lent + interest by that value.

Growth rate = ((money lent + interest)/((1st month * interest rate)+(2nd month * interest rate)+...+(12th month * interest rate))-1)*100

Growth rate = ((100K + 6K)/ (1435.55 + 1112.01 +...+2609.03))-1)*100

Growth rate = 334%

```python
# Interest = last 12 months of Revenue data multiplied by the interest rate
interest_last12months = data_grouped_m.tail(12)['Revenue']*0.06
interest_last12months
```
